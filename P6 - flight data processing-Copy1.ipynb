```python
import sys
import re
import numpy as np
import pandas as pd
import time
from threading import Thread
```

```python
keepList = ['FlightDate', 'DayOfWeek', 
            'Carrier', 'TailNum', 'FlightNum', 
            'OriginState', 'OriginCityName', 'Origin', 
            'DestState', 'Dest', 'DestCityName',
            'DepDelay', 'DepDelayMinutes', 'TaxiOut',
            'ArrDelay', 'ArrDelayMinutes', 'TaxiIn', 
            'Cancelled', 'CancellationCode', 'Diverted', 
            'CRSElapsedTime', 'ActualElapsedTime', 'AirTime', 'Distance', 
            'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
```

```python
# load the dataset
filename = 'On_Time_On_Time_Performance_2016_{0}.csv'
df = pd.read_csv(filename.format(1))
df = df[keepList]

for i in range(2, 13):
    dfTemp = pd.read_csv(filename.format(i))
    dfTemp = dfTemp[keepList]
    df = df.append(dfTemp)
df.reset_index(drop= True, inplace= True)
df['index'] = df.index

# add path_ID columns, warrning, it take 2300 seconds cpu time on my surface pro4 i5 machine
# I tried to use threading module, but in iPython, it doesn't use more than one cpu.
start = time.time()
path_ID = [''.join([df.loc[i, 'Origin'], '_to_', df.loc[i, 'Dest']]) for i in range(df.shape[0])]
end = time.time()
print end - start
df['path_ID'] = path_ID


# save it to Flight_On_Time_Performance_2016.csv
# this file won't be used in Tableau, however, it saves the trouble to regenerate some csv required from Tableau.
filename = 'Flight_On_Time_Performance_2016.csv'
df.to_csv(filename, index = False)
df
```

```python
# prepare path.csv


orderOne = pd.DataFrame(
    {'index': df['index'], 'OriginState': df['OriginState'], 'OriginCity': df['OriginCityName'], 'Airport': df['Origin'],
     'DepDelay': df['DepDelay'], 'DepDelayMinutes': df['DepDelayMinutes'], 'TaxiOut': df['TaxiOut'], 
     'Cancelled': df['Cancelled'], 'CancellationCode': df['CancellationCode'], 
     'Diverted': df['Diverted'], 'path_ID': df['path_ID'],'order': [1 for i in range(df.shape[0])]})
orderTwo = pd.DataFrame(
    {'index': df['index'], 'DestState': df['DestState'], 'DestCity': df['DestCityName'], 'Airport': df['Dest'],
     'ArrDelay': df['ArrDelay'], 'ArrDelayMinutes': df['ArrDelayMinutes'], 'TaxiIn': df['TaxiIn'], 
     'CarrierDelay': df['CarrierDelay'], 'WeatherDelay': df['WeatherDelay'], 
     'NASDelay': df['NASDelay'], 'SecurityDelay': df['SecurityDelay'], 
     'LateAircraftDelay': df['LateAircraftDelay'],
     'path_ID': df['path_ID'], 'order': [2 for i in range(df.shape[0])]})



path = orderOne.append(orderTwo)
path = path.sort_values(by = ['index', 'order'])

filename = 'path.csv'

path.to_csv(filename, index = False)
path
```

```python
# prepare lat, long csv

# load airports lat, lon csv, extract data.
filename = 'airports.csv'
airports = pd.read_csv(filename, header=None)
airports = airports.rename(columns={3: 'Country', 4: 'Airport', 6: 'Latitude', 7: 'Longitude'})
airports = airports[airports['Country'] == 'United States'][['Airport', 'Latitude', 'Longitude']]

# get a unique airport list from path, merge with airports to get lat, long
uniqueAirport = pd.DataFrame(path['Airport'].unique(), columns= ['Airport'])
usaAirport = pd.merge(uniqueAirport, right= airports, on= 'Airport', how= 'left')
usaAirport.index = usaAirport['Airport']

# there are 8 airport whose lat, long are missing, 
fix = {'STT':{'Latitude': 18.3361, 'Longitude': -64.9723}, 
       'SJU':{'Latitude': 18.4374, 'Longitude': -66.0045},
       'STX':{'Latitude': 17.7013, 'Longitude': -64.8058},
       'PSE':{'Latitude': 18.0109, 'Longitude': -66.5632},
       'BQN':{'Latitude': 18.4954, 'Longitude': -67.1356},
       'PPG':{'Latitude': -14.3314, 'Longitude': -170.7114},
       'GUM':{'Latitude': 13.4856, 'Longitude': 144.8001},
       'SPN':{'Latitude': 15.1197, 'Longitude': 145.7283}}

for e in fix:
    usaAirport.set_value(e, 'Latitude', fix[e]['Latitude'])
    usaAirport.set_value(e, 'Longitude', fix[e]['Longitude'])

filename = 'usaAirportsLatLon.csv'
usaAirport.to_csv(filename, index= False)    

usaAirport
```
